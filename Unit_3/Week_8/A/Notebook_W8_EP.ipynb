# Semana 8: Introducción a la Visualización de Datos con Matplotlib y Seaborn

Importar las librerías necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Introducción a Matplotlib
## Crear un gráfico de línea sencillo para representar una tendencia
Crear un gráfico que muestre la tendencia de ventas mensuales de un negocio ficticio.

```python
ventas = np.array([250, 300, 400, 350, 500, 600, 700, 750, 800, 850, 900, 1000])
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

plt.figure(figsize=(10, 6))

plt.plot(meses, ventas, marker='o', linestyle='--', color=(0.3,1,0.8))
plt.title('Tendencia de Ventas Mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas (en unidades)')
plt.grid(True)
plt.show()
```

# Gráfico de Barras con Matplotlib
## Comparar las ventas del primer y segundo semestre del año

```python
ventas_primera_mitad = ventas[:6]
ventas_segunda_mitad = ventas[6:]
semestres = ['Primera Mitad', 'Segunda Mitad']
ventas_totales = [np.mean(ventas_primera_mitad), np.mean(ventas_segunda_mitad)]

print(ventas_totales)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(semestres, ventas_totales, color=['c', 'g'])
plt.title('Comparación de Promedio de Ventas entre los Semestres')
plt.xlabel('Semestres')
plt.ylabel('Ventas Totales (en unidades)')
plt.grid(axis='x')
plt.show()
```

# Gráfico de Dispersión con Matplotlib
## Mostrar la relación entre las ventas mensuales y los meses

```python
plt.figure(figsize=(10, 6))
plt.scatter(meses, ventas,color='m')
plt.title('Relación entre Meses y Ventas')
plt.xlabel('Meses')
plt.ylabel('Ventas (en unidades)')
plt.grid(True)
plt.show()
```

# Ejercicio práctico con Seaborn
 Dataset `tips` incluido en Seaborn
Usar el dataset `tips` de Seaborn para analizar patrones de propinas.

```python
# Configuración básica de estilos para Seaborn
sns.set_theme(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
print(type(tips))

print(tips.head(6))

print("Tamaño: ",tips.size)
```

 Crear un histograma para mostrar la distribución de las propinas

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['tip'], kde=True, color=(0,1,0.5))
plt.title('Distribución de las Propinas')
plt.xlabel('Propina (en USD)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

 Crear un KDEplot para mostrar la densidad de las propinas

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(tips['tip'], color='b')
plt.title('Densidad de las Propinas')
plt.xlabel('Propina (en USD)')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

### Crear un boxplot para analizar diferencias en propinas según el día de la semana

Realizar un boxplot que muestre si existen diferencias significativas en las propinas según el día de la semana.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="tip", hue="sex", data=tips, palette="icefire")
plt.title('Distribución de Propinas según el Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina (en USD)')
plt.grid(True)
plt.show()
```


Crear un violinplot para analizar la distribución y simetría de las propinas

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="day", y="tip", hue="sex", data=tips, palette="viridis")
plt.title('Distribución y Simetría de Propinas según el Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina (en USD)')
plt.grid(True)
plt.show()

```

### Heatmap para mostrar correlaciones

Utilizar un heatmap para visualizar la correlación entre las variables del dataset `tips`.

```python
plt.figure(figsize=(10, 6))
correlacion =tips.select_dtypes(include=['number']).corr()
print(correlacion)
sns.heatmap(correlacion, annot=True, cmap="viridis", linewidths=1)
plt.title('Matriz de Correlación del Dataset `tips`')
plt.show()
```
