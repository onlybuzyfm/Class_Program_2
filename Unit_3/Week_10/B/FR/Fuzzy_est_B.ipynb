### Import Libraries

```python
import numpy as np

import matplotlib.pyplot as plt
```

### Term set Function 

<image src="1.png" >


Universe in range of  $[d \quad f]$

```python
def triangular_mf(x,a,b,c):
    # a inicio del triangulo
    #b el pico 
    # c el final del triangulo
    y= np.maximum(0, np.minimum((x-a)/(b-a),(c-x)/(c-b)))
    return y
```

```python
x=np.linspace(0,10,500)
#print(x)
```

```python
mf1=triangular_mf(x,1,3,5)
mf2 = triangular_mf(x, 4, 6, 8)
mf3 = triangular_mf(x, 7, 9, 10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, mf1, label='Triangular (1, 3, 5)', linewidth=2)
plt.plot(x, mf2, label='Triangular (4, 6, 8)', linewidth=2)
plt.plot(x, mf3, label='Triangular (7, 9, 10)', linewidth=2)
plt.legend()
```

### Caso de Temperatura

```python
x=np.linspace(0,40, 500)
```

```python
cold=triangular_mf(x,1,7,20)
chill = triangular_mf(x, 18, 23, 27)
hot = triangular_mf(x, 22, 30, 36)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x, cold, label='Frio (1, 3, 5)', linewidth=2)
plt.plot(x, chill, label='Templado (4, 6, 8)', linewidth=2)
plt.plot(x, hot, label='Caliente (7, 9, 10)', linewidth=2)
plt.legend()
```

```python
plt.figure(figsize=(10,6))

plt.plot(x, cold, label='Frio (1, 3, 5)', linewidth=2)
plt.plot(x, chill, label='Templado (4, 6, 8)', linewidth=2)
plt.plot(x, hot, label='Caliente (7, 9, 10)', linewidth=2)
plt.legend()

example_points= [19,25,30,32]

for point in example_points:
    y_values=[
        triangular_mf(point, 1,7,20),
        triangular_mf(point,18, 23, 27),
        triangular_mf(point, 22, 30, 36) 
    ]
    print(y_values)
    for i, y in enumerate(y_values):
        plt.plot([point,point], [0,y], 'k--', alpha=0.7)
        # plt.plot([0,y], [point,point],'b--', alpha=0.7)
        plt.text(point, y, f'{y:.2f}', ha='left', va='bottom', fontsize=10)

plt.title('Term Temperature', fontsize=14)
plt.xlabel('Universo de Discurso', fontsize=12)
plt.ylabel('Grado de Membresía', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```
