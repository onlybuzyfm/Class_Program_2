```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define membership functions
def low(x):
    return max(0, min(1, 1 - x / 50))

def high(x):
    return max(0, min(1, (x - 50) / 50))

# Generate data for plotting
x_range = np.linspace(0, 100, 500)
low_values = [low(x) for x in x_range]
high_values = [high(x) for x in x_range]

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x=x_range, y=low_values, label="Low", linewidth=2)
sns.lineplot(x=x_range, y=high_values, label="High", linewidth=2)

# Add labels and title
plt.title("Membership Functions", fontsize=16)
plt.xlabel("Input Value", fontsize=12)
plt.ylabel("Membership Degree", fontsize=12)
plt.legend(title="Fuzzy Sets", fontsize=10)

# Show plot
plt.show()
```

### Defined function in python 

```python
def triangular (a,b,c,x):
    if x<a:
        return 0
    elif a<=x and x<=b:
        return (x-a)/(b-a)
    elif  b<=x and x<=c:
        return abs((x-c)/(c-b))
    else: 
        return 0
    
```

Test function with a=3, b=6, c=9 and x=4

```python

o=triangular(3,6,9,4)
print(o)
```
